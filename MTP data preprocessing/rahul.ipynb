```python
num_devices = 5

# Determine the number of rows per device
rows_per_device = len(new_df) // num_devices

# Generate unique location values for each device ID
# unique_locations = [f'Location_{i}' for i in range(num_devices)]
unique_locations = ['V-3,Vikramshila Complex, IIT Kharagpur, India-721302', 'Kalidas Auditorium, IIT Kharagpur, India-721302','Nalanda Complex, IIT Kharagpur, India-721302','Badminton Court, TSG, IIT Kharagpur, India-721302','Bombay Cineplex, Kharagpur, India-721307']

# Create a list to store the dataframes for each device
device_dfs = []

device_ids = ['1B345','2C578','3D478','4F980','5P345']

# Split the DataFrame into chunks and assign device ID and location
for i in range(num_devices):
    start_idx = i * rows_per_device
    end_idx = start_idx + rows_per_device
    if i == num_devices - 1:  # For the last device, assign the remaining rows
        end_idx = len(new_df)
    device_df = new_df.iloc[start_idx:end_idx].copy()
    device_df['DeviceID'] = device_ids[i]
    device_df['Location'] = unique_locations[i]
    device_dfs.append(device_df)

# Concatenate the dataframes for each device into a single dataframe
result_df = pd.concat(device_dfs, ignore_index=True)
```

```python
cols = list(result_df.columns[-2:]) + list(result_df.columns[:-2])
result_df = result_df[cols]
```

```python
from datetime import datetime, timedelta
start_time = datetime(2024, 4, 10, 0, 0)  # Start time
timestamps = [start_time + timedelta(minutes=i*30) for i in range(len(var_df)//5)]

# Repeat timestamps for each device ID
timestamps = timestamps * len(var_df['DeviceID'].unique())

# Sort the DataFrame by DeviceID
var_df.sort_values(by='DeviceID', inplace=True)

# Assign timestamps to the DataFrame
var_df['Timestamp'] = timestamps
```
